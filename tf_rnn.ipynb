```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential

# 1. 创建模型
model = Sequential([
    layers.Input(shape=(None, 10)),  # 输入形状：时间步长（动态），特征维度：10
    layers.LSTM(64),                 # 64个隐藏单元的LSTM层
    layers.Dense(1)                  # 全连接输出层
])

# 2. 编译模型
model.compile(optimizer='adam', loss='mse')

# 3. 查看网络结构
model.summary()
```
